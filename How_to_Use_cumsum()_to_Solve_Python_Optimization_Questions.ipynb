```python
import pandas as pd

numbers = pd.DataFrame(range(10))
numbers[ 'Cumulative Sum' ] = numbers.cumsum()
```

```python
numbers
```

```python
import pandas as pd

numbers = pd.DataFrame(range(10))
numbers['Cumulative Sum'] = numbers.cumsum()
numbers['reversed'] = list(reversed(range(10)))
numbers['Cumulative Sum 2'] = numbers['reversed'].cumsum()
numbers
```
